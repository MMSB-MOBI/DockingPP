```python
import sys
sys.path.append("/Users/chilpert/Dev/DockingPP/src")
import DockingPP
```

```python
complexes = "/Users/chilpert/Data/Docking/listBM.txt"
results = "/Users/chilpert/Data/Docking/decoys_bm4_zd3.0.2_6deg_fixed/results"
pdb = "/Users/chilpert/Data/Docking/decoys_bm4_zd3.0.2_6deg_fixed/input_pdbs"
results_suffix = ".zd3.0.2.fg.fixed.out"
rmsd_suffix = ".zd3.0.2.fg.fixed.out.rmsds"
ligand_suffix = "_l_u.pdb"
receptor_suffix = "_r_u.pdb"
```

```python
def load_rmsd(rmsd_file, nb_poses):
    rmsds = []
    i = 0
    with open(rmsd_file) as f:
        i += 1
        for l in f:
            rmsds.append(float(l.split("\t")[1]))
        if i == nb_poses:
            return rmsds
            
    return rmsds
```

```python
def countNative(poses, rmsds, rmsd_cutoff):
    nb_native = 0
    for p in poses:
        if rmsds[p.index - 1] <= rmsd_cutoff:
            nb_native += 1
    return nb_native
```

```python
i = 0
success_contact_sum = 0
success_combined = 0
with open(complexes) as complex_list:
    for compl in complex_list:
        i += 1
        compl = compl.rstrip()
        DH = DockingPP.loadZdock(results + "/" + compl + results_suffix, 2000)
        DH.setReceptor(pdb + "/" + compl + receptor_suffix)
        DH.setLigand(pdb + "/" + compl + ligand_suffix)
        DH.computeContactMap(8, 2000)
        DH.computeFrequencies(50)
        DH.rescorePoses(2000, "contacts_sum")
        rmsds = load_rmsd(results + "/" + compl + rmsd_suffix, 2000)
        top10_original = DH.getRankedPoses("original", 10)
        top10_contact = DH.getRankedPoses("contacts_sum", 10)
        contact_natives = countNative(top10_contact, rmsds, 2.5)
        combined_original_contact_55 = top10_original[:5] + [p for p in top10_contact if p not in top10_original][:5]
        combined_natives = countNative(combined_original_contact_55, rmsds, 2.5)
        
        DH.clusterPoses("original", 8, 2000)
        DH.clusterPoses("contacts_sum", 8, 2000)
        DH.pouet()
            
print(success_contact_sum)
print(success_combined)
```
